```python
import requests
import numpy as np
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
df = pd.DataFrame({'job_name':[],'responsibilities':[],'requirements':[],'accessibility':[]})
```

```python
url = "https://ca.indeed.com/viewjob?jk=b56ea92310c792e1&tk=1fs8pj69qsa52800&from=serp&vjs=3"
r = requests.get(url)
soup = bs(r.text,'html.parser')
```

```python
job_name = soup.find('h1').text
```

```python
print(job_name)
```

```python
company = soup.find('div', class_="jobsearch-CompanyReview--heading").text
```

```python
print(company)
```

```python
responsibilities = soup.find('ul')
print(responsibilities)
```

```python
requirements = responsibilities.find_parents()
print(requirements)
```
